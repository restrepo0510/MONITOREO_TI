Se identificó un evento atípico en el que DV_pressure alcanza 9.844 bar. Aunque inicialmente podría interpretarse como una anomalía lógica debido a que COMP = 0 y la corriente del motor es aproximadamente 6 A, la definición de la variable COMP indica que el valor 0 corresponde a que el compresor está cargando. Además, el contexto de presiones cercanas a 10 bar y la activación de DV_electric = 1 sugieren que se trata de un evento transitorio de purga del secador y no de un error de sensor. Por tanto, no es un dato corrupto, sino un evento operativo de alta intensidad que podría ser relevante para la detección temprana de fallas.
